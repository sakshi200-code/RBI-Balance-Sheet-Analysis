## RBI Balance Sheet Basics

Reserve money is the monetary base. It is represented through liability-side components such as currency in circulation and bankers' deposits, and source-side items such as net foreign exchange assets and claims on government.

```python
raw = load_raw()
cleaned = clean_rbi_data()
print(raw.shape)
print(cleaned.shape)
cleaned.head()
```

```python
variable_dictionary = pd.DataFrame({'variable': BALANCE_SHEET_SERIES, 'description': [v.replace('_', ' ').title() for v in BALANCE_SHEET_SERIES]})
variable_dictionary
```

```python
missing = cleaned[BALANCE_SHEET_SERIES].isna().sum().sort_values(ascending=False)
missing.to_frame('missing_values')
```

```python
fig, ax = plt.subplots(figsize=(12, 4))
sns.heatmap(cleaned[BALANCE_SHEET_SERIES].isna(), cbar=False, ax=ax)
ax.set_title('Missing Value Heatmap')
save_figure(fig, '01_missing_value_heatmap.png')
```
