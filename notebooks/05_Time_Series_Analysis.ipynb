# Time Series Analysis

**Objective:** Study trend, seasonality, stationarity, autocorrelation, volatility, and potential structural changes.


```python
import sys
from pathlib import Path

PROJECT_ROOT = Path.cwd().parent if Path.cwd().name == 'notebooks' else Path.cwd()
sys.path.append(str(PROJECT_ROOT / 'src'))

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from rbi_utils import *

sns.set_theme(style='whitegrid', context='notebook')
pd.set_option('display.max_columns', 120)
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.stattools import adfuller, kpss
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

df = load_cleaned().set_index('date').asfreq('W-FRI')
series = df['reserve_money'].interpolate()
```

```python
decomp = seasonal_decompose(series, model='additive', period=52)
fig = decomp.plot()
fig.set_size_inches(12, 8)
save_figure(fig, '05_reserve_money_decomposition.png')
```

```python
adf_result = adfuller(series.dropna())
try:
    kpss_result = kpss(series.dropna(), regression='c', nlags='auto')
except Exception as exc:
    kpss_result = str(exc)
print('ADF statistic:', adf_result[0], 'p-value:', adf_result[1])
print('KPSS:', kpss_result)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 4))
plot_acf(series.dropna(), lags=60, ax=axes[0])
plot_pacf(series.dropna(), lags=60, ax=axes[1], method='ywm')
save_figure(fig, '05_acf_pacf.png')
```
