# Exploratory Data Analysis

**Objective:** Analyze individual trends, distributions, seasonality, and component behavior across the RBI balance sheet.


```python
import sys
from pathlib import Path

PROJECT_ROOT = Path.cwd().parent if Path.cwd().name == 'notebooks' else Path.cwd()
sys.path.append(str(PROJECT_ROOT / 'src'))

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from rbi_utils import *

sns.set_theme(style='whitegrid', context='notebook')
pd.set_option('display.max_columns', 120)
```

```python
df = load_cleaned()
fig, ax = plt.subplots(figsize=(13, 5))
for col in KEY_SERIES:
    ax.plot(df['date'], df[col], label=col.replace('_', ' ').title())
ax.set_title('Key RBI Balance Sheet Series')
ax.legend()
save_figure(fig, '03_key_series_trends.png')
```

```python
fig, axes = plt.subplots(3, 2, figsize=(14, 10))
for ax, col in zip(axes.ravel(), KEY_SERIES):
    sns.histplot(df[col], kde=True, ax=ax)
    ax.set_title(col.replace('_', ' ').title())
fig.delaxes(axes.ravel()[-1])
save_figure(fig, '03_distributions.png')
```

```python
monthly = df.groupby('month_name')[KEY_SERIES].mean().reindex(['January','February','March','April','May','June','July','August','September','October','November','December'])
monthly
```

```python
fig, ax = plt.subplots(figsize=(12, 5))
sns.boxplot(data=df, x='quarter', y='reserve_money', ax=ax)
ax.set_title('Reserve Money by Quarter')
save_figure(fig, '03_quarterly_boxplot.png')
```
