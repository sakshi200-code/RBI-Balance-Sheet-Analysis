# Regression and Correlation

**Objective:** Estimate simple and multiple regression models to identify statistical drivers of reserve money.


```python
import sys
from pathlib import Path

PROJECT_ROOT = Path.cwd().parent if Path.cwd().name == 'notebooks' else Path.cwd()
sys.path.append(str(PROJECT_ROOT / 'src'))

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from rbi_utils import *

sns.set_theme(style='whitegrid', context='notebook')
pd.set_option('display.max_columns', 120)
```

```python
import statsmodels.api as sm

df = load_cleaned().dropna(subset=['reserve_money', 'net_foreign_exchange_assets', 'claims_on_government_net', 'claims_on_banks_and_commercial_sector', 'currency_in_circulation'])
```

```python
X = sm.add_constant(df[['net_foreign_exchange_assets']])
y = df['reserve_money']
simple_model = sm.OLS(y, X).fit()
print(simple_model.summary())
```

```python
features = ['net_foreign_exchange_assets', 'claims_on_government_net', 'claims_on_banks_and_commercial_sector', 'currency_in_circulation']
X_multi = sm.add_constant(df[features])
multi_model = sm.OLS(y, X_multi).fit()
print(multi_model.summary())
```

```python
residuals = multi_model.resid
fig, ax = plt.subplots(figsize=(10, 4))
ax.scatter(multi_model.fittedvalues, residuals, alpha=0.6)
ax.axhline(0, color='black', linewidth=1)
ax.set_title('Multiple Regression Residuals')
save_figure(fig, '06_residuals.png')
```
